## Making figures with numpy arrays

```python
#creating some data
x=np.linspace(0, 10,100)
x[:10]
```

```python
#plotting the data and create a lineplot
fig,ax = plt.subplots()
y = [i**2 for i in x]
ax.plot(x, y);
```

```python
#scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
#scatter sin graph
fig, ax = plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
#making a plot from dictionary
nut_butter_prices = {"Almond butter": 10,
                    "Cashew butter": 8,
                    "Peanut Butter": 12}
fig,ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values());
```

```python
fig,ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()));
```

```python
#histogram
x= np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

## two options for subplots

```python
#subplot option1
fig, ((ax1, ax2), (ax3,ax4)) = plt.subplots(nrows =2,
                                           ncols = 2,
                                           figsize=(10,5))
ax1.plot(x,x/2)
ax2.scatter(x,np.exp(x))
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax4.hist(np.random.randn(100));
```

```python
#subplot option 2
fig, ax = plt.subplots(nrows =2, ncols = 2, figsize=(10,5))
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(x,x);
ax[1,1].hist(np.random.random(100));
```

## Which one you should use? (pyplot vs matplotlib OO method)
* When plotting something quickly, okay to use pyplot method
* when plotting something more advanced, use the OO method

```python
over50 = heart[heart["age"] > 50]
over50.head()
```

```python
#Pyplot method
over50.plot(kind="scatter", x='age', y='chol', c='target');
```

```python
#OO Method mixed with pyplot
fig, ax = plt.subplots(figsize=(10,6))
over50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax);

# ax.set_xlim([45,100])
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
## OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

##plot the data 
scatter = ax.scatter(x=over50["age"], y=over50['chol'], c=over50['target'], cmap="winter")

## Customize the cell
ax.set(title="Heart disease and cholestrol level", xlabel="Age", ylabel="Cholestrol");

## Add legends
ax.legend(*scatter.legend_elements(), title="Target");

## add a horizontal line
ax.axhline(over50["chol"].mean(), linestyle="--");
```

```python
import matplotlib.pyplot as plt

## subplots of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))

## plot the data
scatter0 = ax0.scatter(x=over50["age"], y=over50['chol'], c=over50['target'],cmap="winter")
scatter1 = ax1.scatter(x=over50["age"], y=over50['thalach'], c=over50['target'], cmap="winter")

## customize
ax0.set(title="Heart disease and cholesterol level", xlabel="Age", ylabel="Cholesterol")
ax1.set(title="Heart disease and Thalach level", xlabel="Age", ylabel="Thalach")

## add a horizontal line
ax0.axhline(over50["chol"].mean(), linestyle="--")
ax1.axhline(over50["thalach"].mean(), linestyle="--")

## Add legends
ax0.legend(*scatter0.legend_elements(), title="Target")
ax1.legend(*scatter1.legend_elements(), title="Target")

## Add titel to the fig
fig.suptitle("Heart Disease Analysis", fontsize="16", fontweight="bold");

plt.show()
df["a"].plot()
plt.savefig("images/heart-disease-analysis.png")
```
