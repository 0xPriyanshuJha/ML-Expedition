## Manipulating datas with pandas

```python
import numpy as np
import pandas as pd
```

```python
car_sales = pd.read_csv("car-sales.csv")
```

```python
car_sales
```

```python
car_sales["Make"].str.lower()
```

```python
car_sales["Make"].str.upper()
```

```python
car_sales["Make"] = car_sales["Make"].str.lower()
```

```python
car_sales
```

```python
missing_file = pd.read_csv("car-sales-missing-data.csv")
missing_file
```

```python
# we will mean out the data to fill the missing values
missing_file["Odometer"].fillna(missing_file["Odometer"].mean())
```

```python
missing_file
```

```python
missing_file["Odometer"] = missing_file["Odometer"].fillna(missing_file["Odometer"].mean())
```

```python
missing_file
```

```python
missing_file.dropna(inplace=True)
```

```python
missing_file
```

```python
missing_file_dropped = missing_file.dropna()
```

```python
missing_file_dropped
```

```python
missing_file_dropped.to_csv("car-sales-missing-dropped.csv")
```

```python
car_sales
```

```python
seat_column = pd.Series(['5','4','3','5'])
car_sales['Seats'] = seat_column
car_sales
```

```python
car_sales['Seats'].fillna(5)
```

```python
car_sales
```

```python
fuel_economy = [7.5, 8.5, 6.5, 9.5, 12.5, 6.5, 5.5, 5.0, 8.5, 10.5]
car_sales["Fuel per 100 KM"] = fuel_economy
car_sales
```

```python
car_sales["Total Fuel Used"] = car_sales["Odometer (KM)"]/100 * car_sales["Fuel per 100 KM"]
car_sales
```

```python
car_sales.rename(columns={"Total Fuel Used": "Total Fuel Used (L)"}, inplace=True)
```

```python
car_sales
```

```python
car_sales_shuffled = car_sales.sample(frac=1)
car_sales_shuffled
```

```python
car_sales_shuffled.reset_index(drop=True, inplace=True)
car_sales_shuffled
```

```python
car_sales["Odometer (KM)"] = car_sales["Odometer (KM)"].apply(lambda x:x / 1.6)
car_sales
```
